```python
data9 = [
    ("IND","PAK","IND"),
    ("IND","SL","IND"),
    ("SL","PAK","SL")
]
DF10 = spark.createDataFrame(data9,["team_A","team_B","win"])
DF10.show()                             
```

```python
DF10_1 = DF10.select(col("team_A").alias("team_B"),col("team_B").alias("team_A"),col("win"))
DF10_2 = DF10_1.unionByName(DF10)
DF10_2.show()
DF10_3 = DF10_2.groupBy(col("team_A")).agg(count('*').alias("match_count"))
DF10_3.show()
DF10_4 = DF10.groupBy(col("win")).agg(count("*").alias("win_count"))
DF10_4.show()
DF10_5 = DF10_3.join(DF10_4,DF10_3.team_A==DF10_4.win,"left").select(col("team_A").alias("team"),col("match_count"),col("win_count"))
DF10_5.show()
```

```python
Sample_String = "apple banana orange apple banana graph banana"
DF13 = spark.createDataFrame([(Sample_String,)],["text"])
DF13.show(truncate=False)
```

```python
DF13_New = DF13.withColumn("split",split(col("text")," "))
DF13_New.show(truncate = False)
```

```python
data12 =[(1,"Chemistry",91),
         (1,"Physics",91),
         (1,"Maths",92),
         (2,"Chemistry",80),
         (2,"Physics",90),
         (3,"Chemistry",80),
         (3,"Maths",80),
         (4,"Chemistry",71),
         (4,"Physics",54),
         (5,"Chemistry",79)]
DF14 = spark.createDataFrame(data12,["subject_id","subject","marks"])
DF14.show()         
```

```python
client_date = [("2020-01-02",3),("2020-07-01",7),("2021-01-01",3),("2021-02-03",19),("2022-12-01",3),("2022-12-15",3),("2022-02-28",12)]

client_table = spark.createDataFrame(client_date,["business_date","client_id"])
client_table.show()
```
